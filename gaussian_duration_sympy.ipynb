```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
stim_dur, pref_dur, sigma_dur = sp.symbols('stim_dur pref_dur sigma_dur', real=True)

gaussian = (1 / (sp.sqrt(2 * sp.pi) * sigma_dur)) * \
            sp.exp(- (stim_dur - pref_dur)**2 / (2*sigma_dur**2))
```

```python
# without normalization constant

gaussian_numeric = gaussian.subs({pref_dur: 1.0, sigma_dur: 0.2})

sp.plot(
    gaussian_numeric,
    (stim_dur, 0, 2),
    title="Gaussian Distribution",
    xlabel="Duration",
    ylabel="Probability Density",
    )
```

```python
# Matplotlib

gaussian_func = sp.lambdify(stim_dur, gaussian_numeric, 'numpy')

# Generate values
durations = np.linspace(0,2, 500)
values = gaussian_func(durations)

# Plot
plt.plot(durations, values)
plt.xlabel("Duration")
plt.ylabel("Probability Density")
plt.title("Gaussian Distribution")
plt.show()
```
